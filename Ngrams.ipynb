## Scrape Wikipedia

```python
import bs4 as bs
import urllib.request 

html = urllib.request.urlopen('https://en.wikipedia.org/wiki/tennis')
html = html.read()

article = bs.BeautifulSoup(html, 'lxml')
paragraphs = article.find_all('p')

text = ''
for paragraph in paragraphs:
    text += paragraph.text

text = text.lower()
```

## Only keep letters, periods, and spaces

```python
import re

text = re.sub(r'[^A-Za-z. ]', '', text)

print(text)
```

## Characters N-Grams
* 4-grams

```python
ngrams = {}
n = 4

# iterate starting from (n+1)th character
for i in range(len(text)-n):
    seq = text[i:i+n] # pick next ngram
    if seq not in ngrams.keys(): # verify that it is not already present
        ngrams[seq] = [] # assign empty list to the ngram
    ngrams[seq].append(text[i+n]) # append the character after the ngram as value
    
print(ngrams)
```

**Generate a text of 200 characters**

```python
import random

curr_seq = text[0:n] # store the first n gram
output = curr_seq

for i in range(200): # generate 200 characters
    if curr_seq not in ngrams.keys(): # check if seq in ngrams dict
        break
    poss_chars = ngrams[curr_seq] # list of possible next chars
    next_char = poss_chars[random.randrange(len(poss_chars))] # pick one randomly
    output += next_char
    curr_seq = output[len(output)-n:len(output)] # update curr seq with next ngram
    
print(output)
```

## Words N-Grams
* word 4-grams

```python
import nltk

ngrams = {}
n = 4

tokens = nltk.word_tokenize(text)

for i in range(len(tokens)-n):
    seq = ' '.join(tokens[i:i+n])
    if seq not in ngrams.keys():
        ngrams[seq] = []
    ngrams[seq].append(tokens[i+n])

print(ngrams)
```

**Generate 50 words using first trigram**

```python
curr_seq = ' '.join(tokens[0:n]) # store the first n gram
output = curr_seq

for i in range(50):
    if curr_seq not in ngrams.keys():
        break
    poss_words = ngrams[curr_seq]
    next_word = poss_words[random.randrange(len(poss_words))]
    output += ' '+next_word
    seq_words = nltk.word_tokenize(output)
    curr_seq = ' '.join(seq_words[len(seq_words)-n:len(seq_words)])

print(output)
```
